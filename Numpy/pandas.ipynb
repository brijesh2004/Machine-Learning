# dropna & fillna

```python
import pandas as pd
```

```python
csv_1 = pd.read_csv("C:\\Users\\brsku\\100 Days of ML\\Test_new1.csv")
csv_1
```

```python
csv_1.dropna()   # drop all the value which has NAN value.
# by default it drop with the row wisw
```

```python
csv_1.dropna(axis=1)  # drop the data with the columns wise
```

```python
csv_1.dropna(how="any") # it drop all the row which has any NAN value
csv_1.dropna(how="all")  # it drop the row which has all the value NAN 
```

```python
csv_1.dropna(subset=["d"])  # it remove the row which has NAN in the coloum d 
```

```python
csv_1.dropna(inplace=True)
csv_1
```

```python
csv_1.dropna(thresh=1)   # it remove the row which has exactly 1 NAN value
```

# Fillna

```python
csv_1.fillna("python")  # it replace all the NAN value with the python (that is given in the parenthesis)
```

```python
csv_1.fillna({"a":"python", "s":"c","d":"cpp"})  # if NAN present in the a columns the it replace with the the python
#if the NAN present in the s columns the it replace with the c and with the d columns it replace with the cpp 
```

```python
csv_1.fillna(method="ffill")  # ffill = forward fill if the data is NAN the the previous data is put in that place
```

```python
csv_1.fillna(method="bfill") # backward filling fill with the next data
```

```python
csv_1.fillna(method="ffill",axis=0) # row along 
```

```python
csv_1.fillna(method="ffill" ,axis=1)  # colums along
```

```python
csv_1.fillna(12 , inplace=True) # fill the NAN with the 12
csv_1  
```

```python
csv_1.fillna("python" , limit=1) # it fill the first NAN with the python for all the colums
```

# pivot()

```python
var = pd.DataFrame({"days":[1,2,3,4,5,6] ,"st_name":["a","b","c","a","b","c"], "eng":[10,12,14,15,16,12],"math":[17,18,19,13,14,16]})
var
```

```python
var.pivot(index="days",columns="st_name")
```

```python
var.pivot(index="days",columns="st_name", values="eng")
```

```python
var.pivot_table(index="st_name" , columns = "days" , aggfunc="mean")
```

```python
var.pivot_table(index="st_name" , columns = "days" , aggfunc="sum")
```

```python
var.pivot_table(index="st_name" , columns = "days" , aggfunc="sum",margins=True)
```
